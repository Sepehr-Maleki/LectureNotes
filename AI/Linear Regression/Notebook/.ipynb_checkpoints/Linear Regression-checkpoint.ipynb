## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("Sepehr")
```

## Data

$y(x) = 7x^3 - 3x^2 +2$

```python
def y(x):
    return  7*x**3 - 3*x**2 + 2
```

```python
x = np.random.choice(np.linspace(-2, 2), 35)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(np.linspace(-2, 2), y(np.linspace(-2, 2)), label=r'$y(x) = 7x^3 - 3x^2 +2$', alpha=1, linestyle='dashed', color='gray')
plt.scatter(x, y(x), color='r', label='Data')
plt.legend()
plt.xlim([-2.0005, 2.0050])
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.grid()
plt.tight_layout()
plt.savefig('regintro.eps')
plt.show()
```

## Polynomial Fitting

```python
# Create a dictionary to store model weights for various polynomial degrees
W={}
```

```python
for M in range(0, 18):
    W[M] = np.polynomial.Polynomial.fit(x, y(x), M)
```

```python
def polypred(X, W):
    n = len(W)
    W = np.flip(W.coef)
    S = 0
    l=[]
    for x in X:
        for i in range(1, n):
            S =+ x**i * W[i]
        l.append(S)
    return np.array(l)
```

```python
m=0
n=0
fig, ax = plt.subplots(3, 4, figsize=(18 ,12))
for i in range(1, 13):
    ax[m, n].plot(np.linspace(-2, 2), y(np.linspace(-2, 2)), label=r'$y(x)$', linestyle='dashed', color='black')
    ax[m, n].plot(np.linspace(-2, 2), polypred(np.linspace(-2, 2), W[i]), label='$f(x)$ of degree {0}'.format(i))
    ax[m, n].scatter(x, y(x), color='r', label='Data')
    ax[m, n].legend()
    ax[m, n].set_xlabel('$x$')
    n+=1
    if n >3:
        m+=1
        n=0
plt.tight_layout()
#plt.savefig('polydegrees.eps')
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x, y(x), color='r', label='Data')
plt.plot(np.linspace(-2, 2), polypred(np.linspace(-2, 2), W[1]), label='$f(x)$')
plt.annotate(s=r'', xy=(x[5],y(x[5])+.1), xytext=(x[5],y(x[5])+29), arrowprops=dict(arrowstyle='<->', color='black', linewidth=1.5))
plt.text(x[5]+.1,y(x[5]), r"$e_i = y_i - f(x_i)$")
plt.legend()
plt.tight_layout()
#plt.savefig("err.eps")
```

## Least Squares Fitting

Let's fit a polynomial of degree 3 to the data.

$f(x) = w_3x^3 + w_2 x^2 + w_1 x + w_0$

```python
n = len(x) # number of training examples
m = 4 # order
```

```python
Y = y(x)
```

```python
phi = np.zeros([n,m])
for i in range(len(x)):
    phi[i] = np.array([x[i]**j for j in range(0, m)])
```

```python
w = np.dot(np.dot(np.linalg.inv(np.dot(np.transpose(phi), phi)), np.transpose(phi)), Y)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(x, y(x), color='r', label='Data')
plt.plot(np.linspace(-2, 2, 100), [w[3]*x**3 + w[2]*x**2+w[1]*x + w[0] for x in np.linspace(-2, 2, 100)], label=r'$f(x) = w_3x^3 + w_2 x^2 + w_1 x + w_0$')
plt.xlim([-2.0005, 2.0050])
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.tight_layout()
plt.legend()
```
