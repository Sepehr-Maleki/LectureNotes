```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#plt.style.use('Sepehr')
```

## Data

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3*X + np.random.randn(100, 1)
```

### Basis Function Expansion

$\Phi(x) = \begin{bmatrix}x^3 & x^2 & x & 1\end{bmatrix}$

```python
Phi = np.c_[np.ones_like(X) , X, X**2, X**3]
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(X, y)
```

## Cost Function

```python
def J(w, X, y):
    predictions = X.dot(w)
    cost = np.sum(np.square(predictions - y))
    return cost
```

## Gradient Descent

```python
def gradient_descent(X, y, learning_rate=0.001, iterations=100):
    w = np.random.randn(X.shape[1], 1)
    cost_history = np.zeros(iterations)
    w_history = np.zeros((iterations, w.shape[0]))
    for i in range(iterations):
        prediction = np.dot(X, w)
        w = w - learning_rate * (X.T.dot(prediction - y))
        w_history[i, :] = w.T
        cost_history[i] = J(w, X, y)      
    return w, cost_history, w_history
```

```python
w, cost_hist, w_hist = gradient_descent(Phi, y)
```

### Cost function plot

```python
plt.plot(cost_hist)
```
