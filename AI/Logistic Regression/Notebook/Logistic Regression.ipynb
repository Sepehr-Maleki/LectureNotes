## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('Sepehr')
```

## Sigmoid function

$f(x) = \frac{e^x}{1+e^x} = \frac{1}{1+e^{-x}}$

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
ax.plot(np.linspace(-5, 5, 100), sigmoid(np.linspace(-5, 5, 100)), color='r')
left,right = ax.get_xlim()
low,high = ax.get_ylim()
plt.arrow( left, 0, right -left, 0, length_includes_head = True, head_width = 0.04, head_length=.12, color='black' )
plt.arrow( 0, low, 0, high-low, length_includes_head = True, head_width = 0.12, head_length=.04, color='black' )
plt.text(0.2, high-.03, r"$\sigma(x)$", fontsize=14)
plt.text(right-.2, -.065, r"$x$", fontsize=14)
plt.grid(linestyle='dotted', linewidth=.35)
plt.savefig("sigmoid.eps")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(np.linspace(0, 1, 100), -np.log(sigmoid(np.linspace(0, 1, 100))), color='r')
left,right = ax[0].get_xlim()
low,high = ax[0].get_ylim()
ax[0].arrow( left, 0, right -left, 0, length_includes_head = True, head_width = 0.04, head_length=.1, color='black' )
ax[0].arrow( 0, low, 0, high-low, length_includes_head = True, head_width = 0.1, head_length=.04, color='black' )
ax[0].text(0.2, high-.16, r"$-\log(\sigma(x))$", fontsize=14)
ax[0].text(right-.2, -.285, r"$x$", fontsize=14)
ax[0].grid(linestyle='dotted', linewidth=.35)
ax[0].set_ylim([-0.01, .701])



ax[1].plot(np.linspace(0, 1, 100), -np.log(1 - sigmoid(np.linspace(0, 1, 100))), color='r')
left_,right_ = ax[1].get_xlim()
low_,high_ = ax[1].get_ylim()
ax[1].arrow(left_, 0, right_ -left_, 0, length_includes_head = True, head_width = 0.1, head_length=.02, color='black' )
ax[1].arrow(0, high_, 0, low_, length_includes_head = True, head_width = 0.02, head_length=.1, color='black' )
ax[1].text(0.02, low_+.16, r"$-\log(1-\sigma(x))$", fontsize=14)
ax[1].text(right_-.02, -.25, r"$x$", fontsize=14)
ax[1].grid(linestyle='dotted', linewidth=.35)
ax[1].set_ylim([-.5, 1.5])
plt.subplots_adjust(right = 1.5)
#plt.savefig("logcost.eps", bbox_inches='tight')
```
