## Imports

```python
from scipy.stats import beta, binom
from scipy.special import binom as binom_coef
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

## Example

Suppose we're given a population where a fraction of people are suffering from a specific disease. We draw a sample of size $N=10$ (select 10 people from the population) to perform some blood tests. The results reveal that only 1 person has the disease. We would like to know what is the probability that any person from the population has the disease. In fact, if we denote the parameter that governs this probability with $\theta$, we want to find what posterior distribution we can assign to $\theta$ after observing our sample of 10 people.

### Likelihood

It is easy to see that a binomial distribution is an appropriate choice for likelihood:

$p(X|\theta) = {10\choose 1}~ \theta^1~ (1-\theta)^9$

```python
N = 10
k = 1
```

```python
coef = binom_coef(N, k)
```

Binomial PMF: ${N \choose k} \theta^k (1-\theta)^{N-k}$

```python
likelihood = [coef * theta**k * (1-theta)**(N-k) for theta in np.linspace(0, 1, 100)]
```

### Prior

Let's assume that we don't really have any idea as to what percentage of the population is suffering from the disease. In this case a uniform distribution seems to be an appropriate choice to represent our ``non-informative`` prior. However, instead, we use a Beta distribution which can be equivalent to a uniform distribution if $Beta(1, 1)$. Therefore:

$p(\theta) \sim Beta(1, 1)$

```python
plt.figure(figsize=(8, 8))
plt.plot(np.linspace(0, 1, 100), [beta(1, 1).pdf(x) for x in np.linspace(0, 1, 100)], linewidth=1)
plt.title("Beta Distribution PDF")
plt.tight_layout()
plt.savefig("betauniform.eps")
```

```python
prior = beta(1, 1)
```

### Posterior

$p(\theta | X) = \frac{p(X|\theta) \times p(\theta)}{p(X)}$

One way to find the posterior distribution, i.e., $p(\theta | X)$ is to find the above numerator and then integrate it to find the denominator. However, we don't need to! Since we have a Beta prior, which is conjugate to the binomial distribution for likelihood, our posterior is also a Beta distribution with the following parameters:

$p(\theta | X) \sim Beta(1+1, 1+10-1)$.

In general if 

$p(\theta) \sim Beta(a, b)$,

and $p(X | \theta)$ has the PMF: ${N \choose k} \theta^k (1-\theta)^{N-k}$,

Then:

$p(\theta|X) \sim Beta(a+b, b+N-X)$.



```python
posterior = beta(2, 10)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(16, 8))
ax[0].plot(np.linspace(0, 1, 100), [prior.pdf(x) for x in np.linspace(0, 1, 100)], linewidth=1)
ax[0].set_title("Prior")

ax[1].plot(np.linspace(0, 1, 100), likelihood, 'o', markersize=1.7, color='green')
ax[1].set_title("Likelihood")

ax[2].plot(np.linspace(0, 1, 100), [posterior.pdf(x) for x in np.linspace(0, 1, 100)], linewidth=1, color='red')
ax[2].set_title("Posterior")
plt.tight_layout()
plt.savefig("dpposterior.eps")
```
