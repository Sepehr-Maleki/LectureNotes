## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
plt.style.use("fivethirtyeight")
```

## Example

Suppose that the posterior distribution for $\theta$ is a Beta$(1, 24)$ distribution, with probability density function:

$p(\theta| data) = 24(1- \theta)^{23}, \quad 0 < \theta < 1\;.$

Determine the $95\%$ HDI for $\theta$.

```python
posterior = beta(1, 24)
```

```python
plt.figure(figsize=(8, 3))
plt.plot(np.linspace(0, 1, 100), [posterior.pdf(theta) for theta in np.linspace(0, 1, 100)], linewidth=1)
plt.title("Posterior")
plt.xlabel(r"$\theta$")
plt.ylabel("Density")
plt.grid(True)
plt.savefig("crposterior.eps")
```

$\alpha = 0.05 \Longrightarrow$ the $95\%$ HDI for $\theta$ is $(0, 1- 0.05^{1/24})$.

```python
a = 0
b = 1 - 0.05**(1/24)
```

```python
plt.figure(figsize=(12, 4))
plt.plot(np.linspace(0, 1, 100), [posterior.pdf(theta) for theta in np.linspace(0, 1, 100)])
plt.axvspan(a, b, alpha=0.1, color='orange', label="$95%$ Credible Interval")
plt.title("Posterior")
plt.xlabel(r"$\theta$")
plt.ylabel("Density")
plt.grid(True)
plt.legend()
plt.savefig("credible.eps")
plt.show()
```
