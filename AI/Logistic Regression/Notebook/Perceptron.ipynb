## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
plt.style.use("Sepehr")
%matplotlib inline
```

## Perceptron

### Data

```python
X, y = make_blobs(n_samples=200, n_features=2, centers=2, random_state=1)
```

```python
y = [1 if e==1 else -1 for e in y]
```

### Vis.

```python
# Define colours
C = ['#008fd5' if x==1 else 'red' for x in y]
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], c=C)
```

### Devide to test/train sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### Perceptron Algorithm

```python
class Perceptron(object):
    
    """
    Class for performing Perceptron.
    X is the input array with n rows (no_examples) and m columns (no_features)
    Y is a vector containing elements which indicate the class 
        (1 for positive class, -1 for negative class)
    w is the weight-vector (m number of elements)
    b is the bias-value
    """
    
    def __init__(self, b=0, max_iter=1000):
        self.max_iter = max_iter
        self.w = []
        self.b = b
        self.n_examples = 0
        self.n_features = 0
        self.W = []
        self.B = []
        
    def train(self, X, y):
        self.n_examples, self.n_features = np.shape(X)
        self.w = np.zeros(self.n_features)
        for i in range(0, self.max_iter):
            w_updated = False
            for e in range(0, self.n_examples):
                a = self.b + np.dot(self.w, X[e])
                if np.sign(y[e]*a) != 1:
                    w_updated = True
                    self.w += y[e] * X[e]
                    self.b += y[e]
                    self.W.append(self.w)
                    self.B.append(self.b)
                    
            if not w_updated:
                print("Convergence reached in %i iterations." % i)
                break
        if w_updated:
            print(
            """
            WARNING: convergence not reached in %i iterations.
            Either dataset is not linearly separable, 
            or max_iter should be increased
            """ % self.max_iter
                )
    
    def classify_element(self, x):
        return int(np.sign(self.b+np.dot(self.w, x)))
    
    def classify(self, X):
        predictions = []
        for e in range(np.shape(X)[0]):
            prediction = self.classify_element(X[e])
            predictions.append(prediction)
        return predictions
    
    def check_accuracy(self, predictions, true_y):
        percentage = sum([predictions[i]==true_y[i] for i in range(len(true_y))])/len(true_y) *100
        print("Accuracy: {0}%".format(percentage))
        
    def calc_hyperplane(self, X):
        return np.ravel([-(self.b + x * self.w[0]) / self.w[1] for x in [X[:, 0].min(), X[:, 1].max()]])
        
    
    def get_params(self):
        return self.w, self.b
    
    def params_hist(self):
        return self.W, self.B
```

```python
P = Perceptron()
```

#### Train

```python
P.train(X_train, y_train)
```

##### Predict

```python
predictions = P.classify(X_test)
```

#### Accuracy

```python
P.check_accuracy(predictions, y_test)
```
