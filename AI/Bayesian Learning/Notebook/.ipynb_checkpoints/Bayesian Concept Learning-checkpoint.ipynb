## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

## Hypotheses

```python
Hypotheses_names = ["mult {0}".format(i) for i in range(3, 11)] +\
["ends {0}".format(i) for i in range(1, 10)] +\
["pow {0}".format(i) for i in range(2, 11)] +\
["even", "odd", "squares", "pow 2 + {37}", "pow 2 - {32}", "All"]
```

```python
Hypotheses = dict.fromkeys(Hypotheses_names)
```

```python
Hypotheses["even"] = [x for x in range(1, 101) if x%2==0]
Hypotheses["odd"] = [x for x in range(1, 101) if x%2==1]
Hypotheses["squares"] = [x**2 for x in range(1, 101) if x**2 <101]

# Multiplications by *
Hypotheses["mult 3"] = [3*x for x in range(1, 101) if 3*x <101]
Hypotheses["mult 4"] = [4*x for x in range(1, 101) if 4*x <101]
Hypotheses["mult 5"] = [5*x for x in range(1, 101) if 5*x <101]
Hypotheses["mult 6"] = [6*x for x in range(1, 101) if 6*x <101]
Hypotheses["mult 7"] = [7*x for x in range(1, 101) if 7*x <101]
Hypotheses["mult 8"] = [8*x for x in range(1, 101) if 8*x <101]
Hypotheses["mult 9"] = [9*x for x in range(1, 101) if 9*x <101]
Hypotheses["mult 10"] = [10*x for x in range(1, 101) if 10*x <101]

# Ends in *
Hypotheses["ends 1"] = [x for x in range(1, 101) if x%10==1]
Hypotheses["ends 2"] = [x for x in range(1, 101) if x%10==2]
Hypotheses["ends 3"] = [x for x in range(1, 101) if x%10==3]
Hypotheses["ends 4"] = [x for x in range(1, 101) if x%10==4]
Hypotheses["ends 5"] = [x for x in range(1, 101) if x%10==5]
Hypotheses["ends 6"] = [x for x in range(1, 101) if x%10==6]
Hypotheses["ends 7"] = [x for x in range(1, 101) if x%10==7]
Hypotheses["ends 8"] = [x for x in range(1, 101) if x%10==8]
Hypotheses["ends 9"] = [x for x in range(1, 101) if x%10==9]

Hypotheses["pow 2"] = [2**x for x in range(1, 101) if 2**x <101]
Hypotheses["pow 3"] = [3**x for x in range(1, 101) if 3**x <101]
Hypotheses["pow 4"] = [4**x for x in range(1, 101) if 4**x <101]
Hypotheses["pow 5"] = [5**x for x in range(1, 101) if 5**x <101]
Hypotheses["pow 6"] = [6**x for x in range(1, 101) if 6**x <101]
Hypotheses["pow 7"] = [7**x for x in range(1, 101) if 7**x <101]
Hypotheses["pow 8"] = [8**x for x in range(1, 101) if 8**x <101]
Hypotheses["pow 9"] = [9**x for x in range(1, 101) if 9**x <101]
Hypotheses["pow 10"] = [10**x for x in range(1, 101) if 10**x <101]

# All numbers
Hypotheses["All"] = [x for x in range(1, 101)]

# Unnatural
Hypotheses["pow 2 + {37}"] = [2**x for x in range(1, 101) if 2**x <101] + [37]
Hypotheses["pow 2 - {32}"] = [2**x for x in range(1, 101) if (2**x <101) and 2**x != 32]
```

## Prior probabilities

Assign some probabilities to the priors for hypotheses

```python
Priors = dict.fromkeys(Hypotheses_names)
```

```python
for key in Priors:
    if (key.startswith("mult")) or (key.startswith("ends")) or (key.startswith("pow")) or (key == "squares") or (key == "All"):
        Priors[key] = .02
    if (key == 'even') or (key == 'odd'):
        Priors[key] = .15
    if (key.endswith("{37}")) or (key.endswith("{32}")):
        Priors[key] = .002
```

```python
plt.figure(figsize=(16, 8))
plt.title("Prior")
plt.bar(range(0, len(Hypotheses)), Priors.values())
plt.xticks(range(len(Hypotheses)), Hypotheses, size='small', rotation=90)
plt.tight_layout()
#plt.savefig("priors.eps")
plt.show()
```

## Likelihood

### D = {16}

```python
Likelihood_16 = dict.fromkeys(Hypotheses_names)
```

```python
for key in Likelihood_16:
    if 16 in Hypotheses[key]:
        Likelihood_16[key] = 1/len(Hypotheses[key])
    else:
        Likelihood_16[key] = 0
```

```python
plt.figure(figsize=(16, 8))
plt.title("Likelihood")
plt.bar(range(0, len(Hypotheses)), list(Likelihood_16.values()))
plt.xticks(range(len(Hypotheses)), Hypotheses, size='small', rotation=90)
plt.tight_layout()
#plt.savefig("Likelihoods.eps")
plt.show()
```

#### Prior-Likelihood plots

```python
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(14, 8))

axes[0].bar(range(0, len(Hypotheses)), Priors.values())
axes[0].set_title("Prior", fontsize=16)
axes[0].grid(True)
axes[0].set_xticks(range(len(Hypotheses)), Hypotheses)

axes[1].bar(range(0, len(Hypotheses)), list(Likelihood_16.values()))
axes[1].set_title("Likelihood", fontsize=16)
axes[1].set_xticks(range(0, len(Hypotheses_names)))
axes[1].set_xticklabels(Hypotheses_names, rotation='vertical', fontsize=10)
plt.tight_layout()
axes[1].grid(True)

#plt.savefig("Prior_Likelihood_16.eps")
```

### D = {16, 8, 2, 64}

```python
D = [16, 8, 2, 64]
N = len(D)
```

```python
Likelihood_D = dict.fromkeys(Hypotheses_names)
```

```python
for key in Likelihood_D:
    if sum([True for x in D if x in Hypotheses[key]]) == N:
        Likelihood_D[key] = (1/len(Hypotheses[key]))**N
    else:
        Likelihood_D[key] = 0
```

```python
plt.figure(figsize=(16, 8))
plt.title("Likelihood")
plt.bar(range(0, len(Hypotheses)), list(Likelihood_D.values()))
plt.xticks(range(len(Hypotheses)), Hypotheses, size='small', rotation=90)
plt.tight_layout()
#plt.savefig("Likelihoods.eps")
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(14, 8))

axes[0].bar(range(0, len(Hypotheses)), Priors.values())
axes[0].set_title("Prior", fontsize=16)
axes[0].grid(True)
axes[0].set_xticks(range(len(Hypotheses)), Hypotheses)

axes[1].bar(range(0, len(Hypotheses)), list(Likelihood_D.values()))
axes[1].set_title("Likelihood", fontsize=16)
axes[1].set_xticks(range(0, len(Hypotheses_names)))
axes[1].set_xticklabels(Hypotheses_names, rotation='vertical', fontsize=10)
plt.tight_layout()
axes[1].grid(True)

#plt.savefig("Prior_Likelihood_D.eps")
```

## Posterior

### D = {16}

```python
Posterior_16 = dict.fromkeys(Hypotheses_names)
```

#### Normalisation Factor (Marginal Likelihood)

```python
NF = 0
for key in Posterior_16:
    NF += Likelihood_16[key] * Priors[key]
```

```python
for key in Posterior_16:
    Posterior_16[key] = (Priors[key] * Likelihood_16[key])/NF
```

```python
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(14, 10))

# Prior
axes[0].bar(range(0, len(Hypotheses)), Priors.values())
axes[0].set_title("Prior", fontsize=16)
axes[0].grid(True)
axes[0].set_xticks(range(len(Hypotheses)), Hypotheses)

# Likelihood
axes[1].bar(range(0, len(Hypotheses)), list(Likelihood_16.values()))
axes[1].set_title("Likelihood", fontsize=16)
axes[1].set_xticks(range(0, len(Hypotheses_names)))
axes[1].grid(True)

# Posterior
axes[2].bar(range(0, len(Hypotheses)), list(Posterior_16.values()))
axes[2].set_title("Posterior", fontsize=16)
axes[2].set_xticks(range(0, len(Hypotheses_names)))
axes[2].set_xticklabels(Hypotheses_names, rotation='vertical', fontsize=10)
axes[2].grid(True)

plt.savefig("numbersGame16.eps", bbox_inches = 'tight')
```

### D = {16, 8, 2, 64}

```python
Posterior_D = dict.fromkeys(Hypotheses_names)
```

#### Normalisation Factor (Marginal Likelihood)

```python
NF = 0
for key in Posterior_D:
    NF += Likelihood_D[key] * Priors[key]
```

```python
for key in Posterior_D:
    Posterior_D[key] = (Priors[key] * Likelihood_D[key])/NF
```

```python
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(14, 10))

# Prior
axes[0].bar(range(0, len(Hypotheses)), Priors.values())
axes[0].set_title("Prior", fontsize=16)
axes[0].grid(True)
axes[0].set_xticks(range(len(Hypotheses)), Hypotheses)

# Likelihood
axes[1].bar(range(0, len(Hypotheses)), list(Likelihood_D.values()))
axes[1].set_title("Likelihood", fontsize=16)
axes[1].set_xticks(range(0, len(Hypotheses_names)))
axes[1].grid(True)

# Posterior
axes[2].bar(range(0, len(Hypotheses)), list(Posterior_D.values()))
axes[2].set_title("Posterior", fontsize=16)
axes[2].set_xticks(range(0, len(Hypotheses_names)))
axes[2].set_xticklabels(Hypotheses_names, rotation='vertical', fontsize=10)
axes[2].grid(True)

plt.savefig("numbersGameD.eps", bbox_inches = 'tight')
```

## Posterior Predictive Distribution

### D = {16}

```python
PP = []
S = 0
for xhat in range(1, 101):
    for h in Hypotheses:
        if xhat in Hypotheses[h]:
            S += Posterior_16[h]
    PP.append(S)
    S=0
```

```python
L = []
for xhat in range(1, 101):
    for h in Hypotheses:
        if xhat in Hypotheses[h]:
            L.append((xhat, h)) 
x = [item[0] for item in L]
y = [item[1] for item in L]
```

```python
fig = plt.figure(figsize=(16, 12))

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
ax1.bar(np.arange(1, 101), PP)
ax1.set_xticks(np.arange(0, 101, 4))
ax1.set_yticks(np.arange(0, 1.1, .5))
ax1.set_title("Posterior Predictive Distribution")

ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2, rowspan=3)
ax2.plot(x, y, 'o')
ax2.set_xticks([])

ax3 = plt.subplot2grid((3, 3), (0, 2), rowspan=3)
ax3.plot(list(Posterior_16.values()), Hypotheses_names, '-o', linewidth=1)
ax3.set_title("$p(h|16)$")

plt.tight_layout()
plt.savefig("posterior_predictive_16.eps")
plt.show()
```

### D = {16, 8, 2, 64}

```python
PP = []
S = 0
for xhat in range(1, 101):
    for h in Hypotheses:
        if xhat in Hypotheses[h]:
            S += Posterior_D[h]
    PP.append(S)
    S=0
```

```python
fig = plt.figure(figsize=(16, 12))

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
ax1.bar(np.arange(1, 101), PP)
ax1.set_xticks(np.arange(0, 101, 4))
ax1.set_yticks(np.arange(0, 1.1, .5))
ax1.set_title("Posterior Predictive Distribution")

ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2, rowspan=3)
ax2.plot(x, y, 'o')
ax2.set_xticks([])

ax3 = plt.subplot2grid((3, 3), (0, 2), rowspan=3)
ax3.plot(list(Posterior_D.values()), Hypotheses_names, '-o', linewidth=1)
ax3.set_title("$p(h|\{16, 8, 2, 64\})$")

plt.tight_layout()
plt.savefig("posterior_predictive_D.eps")
plt.show()
```
