## Imports

```python
import numpy as np
from sklearn.datasets import make_circles, make_classification
from sklearn.model_selection import train_test_split
import tensorflow as tf

import matplotlib.pyplot as plt
%matplotlib inline
#plt.style.use("Sepehr")
```

## Data

```python
X, y = make_classification(n_samples=50000, n_features=2, n_classes=2,
                           n_clusters_per_class=1,
                           n_informative=1, n_redundant=0)
```

### Training/Test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

#### Training dataset

```python
plt.figure(figsize=(16, 8))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='rainbow', s=1)
```

#### Test dataset

```python
plt.figure(figsize=(16, 8))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='rainbow', s=1)
```

### One Hot Encoding

```python
y_train = np.array([y_train, -(y_train-1)]).T
y_test = np.array([y_test, -(y_test-1)]).T
```

## Neural Network

### Parameters

```python
learning_rate = 0.1
num_epochs = 500
batch_size = 128
total_batches = int(len(X_train)/batch_size)
display_step = 100
```

### Network Parameters

```python
n_hidden_1 = 256 # 1st layer number of neurons
n_hidden_2 = 256 # 2nd layer number of neurons
num_features = 2 # number of features
num_classes = 2 # number of classes
```

### tf Graph input

```python
X = tf.placeholder("float", [None, num_features])
Y = tf.placeholder("float", [None, num_classes])        
```

### Weights & Biases

```python
weights = {
    'h1': tf.Variable(tf.random_normal([num_features, n_hidden_1])),
    'h2': tf.Variable(tf.random_normal([n_hidden_1, n_hidden_2])),
    'out': tf.Variable(tf.random_normal([n_hidden_2, num_classes]))
}

biases = {
    'b1': tf.Variable(tf.random_normal([n_hidden_1])),
    'b2': tf.Variable(tf.random_normal([n_hidden_2])),
    'out': tf.Variable(tf.random_normal([num_classes]))
}
```

### Feed-forward Operation

```python
def feedfwd(x):
    # Hidden fully connected layer with 256 neurons
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1 = tf.nn.relu(layer_1)
    # Hidden fully connected layer with 256 neurons
    layer_2 = tf.add(tf.matmul(layer_1, weights['h2']), biases['b2'])
    layer_2 = tf.nn.relu(layer_2)
    # Output fully connected layer with a neuron for each class
    out_layer = tf.matmul(layer_2, weights['out']) + biases['out']
    return out_layer
```

### Learning (Backprop)

```python
# Define logits
logits = feedfwd(X)

# Define cost
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits, labels=Y))

# Define optimiser
optimiser = tf.train.AdamOptimizer(learning_rate=learning_rate)
train_op = optimiser.minimize(cost)


# Evaluate model (with test logits, for dropout to be disabled)
correct_pred = tf.equal(tf.argmax(logits, 1), tf.argmax(Y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

### Training

```python
# Initialize the variables (i.e. assign their default value)
init = tf.global_variables_initializer()

# Start training
with tf.Session() as sess:

    # Run the initializer
    sess.run(init)

    for epoch in range(1, num_epochs):
        X_batches = np.array_split(X_train, total_batches)
        Y_batches = np.array_split(y_train, total_batches)   
        for batch in range(total_batches):
            batch_x, batch_y = X_batches[batch], Y_batches[batch]
            # Run optimisation op (backprop)
            sess.run(train_op, feed_dict={X: batch_x, Y: batch_y})
        if epoch % display_step == 0 or epoch == 1:
            # Calculate batch cost and accuracy
            loss, acc = sess.run([cost, accuracy], feed_dict={X: batch_x,
                                                                     Y: batch_y})
            print("Epoch " + str(epoch) + ", Minibatch Cost= " + \
                    "{:.4f}".format(loss) + ", Training Accuracy= " + \
                    "{:.3f}".format(acc))

    print("Optimisation Finished!")

    # Calculate accuracy for MNIST test images
    print("Testing Accuracy:", \
        sess.run(accuracy, feed_dict={X: X_test,
                                      Y: y_test}))
    p = sess.run(logits, feed_dict={X: X_test[0:10]})
    global result 
    result = tf.argmax(logits, 1).eval({X: X_test, Y: y_test})
```
