## Neural Network

### Activation Function

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def sigmoid_backward(dX, z):
    sig = sigmoid(z)
    return dX * sig * (1-sig)
```

### Forward Propagation

```python
def fwd_prop(X, W, b):
    z = np.dot(W, X) + b
    y = sigmoid(z)
    return y
```

### Loss Function / Metrics

```python
def loss(y, y_true):
    m = len(y)
    cost = -1/m * (np.dot(y_true, np.log(y).T) + np.dot(1 - y_true, np.log(1-y).T))
    return np.squeeze(cost)
```
