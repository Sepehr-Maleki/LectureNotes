## Sigmoid function

$f(x) = \frac{e^x}{1+e^x} = \frac{1}{1+e^{-x}}$

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
ax.plot(np.linspace(-5, 5, 100), sigmoid(np.linspace(-5, 5, 100)), color='r')
left,right = ax.get_xlim()
low,high = ax.get_ylim()
plt.arrow( left, 0, right -left, 0, length_includes_head = True, head_width = 0.04, head_length=.12, color='black' )
plt.arrow( 0, low, 0, high-low, length_includes_head = True, head_width = 0.12, head_length=.04, color='black' )
plt.text(0.2, high-.03, r"$\sigma(z)$", fontsize=14)
plt.text(right-.2, -.065, r"$z$", fontsize=14)
plt.savefig("sigmoid.eps")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(np.linspace(0, 5, 100), np.log(np.linspace(0, 5, 100)), color='r')
left,right = ax[0].get_xlim()
low,high = ax[0].get_ylim()
ax[0].arrow( left, 0, right -left, 0, length_includes_head = True, head_width = 0.15, head_length=.12, color='black' )
ax[0].arrow( 0, low, 0, high-low, length_includes_head = True, head_width = 0.12, head_length=.15, color='black' )
ax[0].text(0.2, high-.16, r"$\log(x)$", fontsize=14)
ax[0].text(right-.2, -.285, r"$x$", fontsize=14)

ax[1].plot(np.linspace(0, 5, 100), np.log(1 - np.linspace(0, 5, 100)), color='r')
left_,right_ = ax[1].get_xlim()
low_,high_ = ax[1].get_ylim()
ax[1].arrow(left_, 0, right_ -left_, 0, length_includes_head = True, head_width = 0.1, head_length=.02, color='black' )
ax[1].arrow(0, high_, 0, low_, length_includes_head = True, head_width = 0.02, head_length=.1, color='black' )
ax[1].text(0.02, low_+.16, r"$\log(1-x)$", fontsize=14)
ax[1].text(right_-.02, -.25, r"$x$", fontsize=14)
plt.subplots_adjust(right = 1.5)
#plt.savefig("logcost.eps", bbox_inches='tight')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(np.linspace(-3, 3, 100), -np.log(sigmoid(np.linspace(-3, 3, 100))), color='r')
left,right = ax[0].get_xlim()
low,high = ax[0].get_ylim()
ax[0].arrow( left, 0, right -left, 0, length_includes_head = True, head_width = 0.15, head_length=.1, color='black' )
ax[0].arrow( 0, low, 0, high-low, length_includes_head = True, head_width = 0.11, head_length=.13, color='black' )
ax[0].text(0.2, high-.05, r"$-\log(\frac{1}{1+e^{-z}})$", fontsize=14)
ax[0].text(right-.1, -.2, r"$z$", fontsize=14)
ax[0].plot(np.linspace(1, 3, 100), np.zeros(100), linewidth=4)
ax[0].plot(np.linspace(-3, 1, 100), -.6*np.linspace(-3, 1, 100)+.6, linewidth=4, color='#008fd5')
ax[0].vlines(1, -.1, .1, linewidth=3)
ax[0].text(-1.7, 1, r'$Cost_1 (z)$', color='#008fd5', fontsize=16)



ax[1].plot(np.linspace(-3, 3, 100), -np.log(1 - sigmoid(np.linspace(-3, 3, 100))), color='r')
left_,right_ = ax[1].get_xlim()
low_,high_ = ax[1].get_ylim()
ax[1].arrow(left_, 0, right_ - left_, 0, length_includes_head = True, head_width = 0.15, head_length=.1, color='black' )
ax[1].arrow( 0, low_, 0, high_ - low_, length_includes_head = True, head_width = 0.11, head_length=.13, color='black' )
ax[1].text(0.07, high_-.05, r"$- \log(1 - \frac{1}{1+e^{-z}})$", fontsize=14)
ax[1].text(right_-.1,-.2, r"$z$", fontsize=14)
ax[1].plot(np.linspace(-3, -1, 100), np.zeros(100), linewidth=4)
ax[1].plot(np.linspace(-1, 3, 100), .6*np.linspace(-3, 1, 100)+1.8, linewidth=4, color='#008fd5')
ax[1].vlines(-1, -.1, .1, linewidth=3)
ax[1].text(1, 1, r'$Cost_0 (z)$', color='#008fd5', fontsize=16)

plt.subplots_adjust(right = 1.5)
plt.savefig("logcostsvm.eps", bbox_inches='tight')
```

## Example

```python
np.random.seed(111)
cross_x = np.arange(1, 9)
np.random.shuffle(cross_x)
cross_y = np.random.choice(range(1, 20), 8, replace=False)

circ_x = np.arange(12, 20)
np.random.shuffle(circ_x)
circ_y = np.random.choice(range(1, 20), 8, replace=False)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

ax.scatter(cross_x, cross_y, marker='+', s=88)
ax.scatter(circ_x, circ_y, marker='o', facecolor='none', edgecolor='red', linewidth=1.5, s=88)
left,right = ax.get_xlim()
low,high = ax.get_ylim()
ax.arrow(-.5, 0, right+.5, 0, length_includes_head = True, head_width = 0.45, head_length=.18, color='black' )
ax.arrow( 0, -1, 0, high+2.5, length_includes_head = True, head_width = 0.2, head_length=.38, color='black' )
ax.plot(np.linspace(5, 14, 100), -4*np.linspace(5, 14, 100)+56.8, linewidth=1, color='green')
ax.plot(np.linspace(5, 12, 100), -12.1*np.linspace(5, 12, 100)+146, linewidth=1, color='m')
ax.vlines(11, 0, 20, color='orange')
plt.ylim([-1, 21])
plt.xlim([-1, 20])
plt.savefig("svmbaddecision.eps")
```

## SVM

```python
X, y = make_blobs(n_samples=10, centers=2, random_state=6)
```

```python
clf = svm.SVC(kernel='linear', C=1000, gamma='auto')
clf.fit(X, y)
```

```python
Y = ['#008fd5' if v==1 else 'red' for v in y]
```

```python
Y
```

```python
M = ['+' if v==1 else 'o' for v in y]
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, s=100)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.plot(np.linspace(5, 12, 100), -1.2*np.linspace(5, 12, 100)+2.5, linewidth=1.5, color='m')
ax.plot(np.linspace(5, 12, 100), -1.6*np.linspace(5, 12, 100)+5, linewidth=1.5, color='g')
ax.plot(np.linspace(5, 12, 100), -2*np.linspace(5, 12, 100)+7.5, linewidth=1.5, color='orange')


# plot decision boundary and margins
# ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
#            linestyles=['--', '-', '--'])
# plot support vectors
#ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
#           linewidth=1, facecolors='none', edgecolors='k')

left,right = ax.get_xlim()
low,high = ax.get_ylim()
ax.arrow(5, -9.5, 6, 0, length_includes_head = True, head_width = 0.3, head_length=.1, color='black' )
ax.arrow( 5.2, -10, 0, 8, length_includes_head = True, head_width = 0.1, head_length=.3, color='black' )

plt.ylim([-10, -2])
plt.xlim([5, 11])
plt.savefig("svmbaddecision.eps")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y, s=100)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.plot(np.linspace(5, 12, 100), -1.2*np.linspace(5, 12, 100)+2.5, linewidth=1.5, color='m', alpha=.3)
ax.plot(np.linspace(5, 12, 100), -1.6*np.linspace(5, 12, 100)+5, linewidth=1.5, color='g', alpha=.3)
ax.plot(np.linspace(5, 12, 100), -2*np.linspace(5, 12, 100)+7.5, linewidth=1.5, color='orange', alpha=.3)


# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support vectors
#ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
#           linewidth=1, facecolors='none', edgecolors='k')

left,right = ax.get_xlim()
low,high = ax.get_ylim()
ax.arrow(5, -9.5, 6, 0, length_includes_head = True, head_width = 0.3, head_length=.1, color='black' )
ax.arrow( 5.2, -10, 0, 8, length_includes_head = True, head_width = 0.1, head_length=.3, color='black' )

plt.ylim([-10, -2])
plt.xlim([5, 11])
plt.savefig("svmgooddecision.pdf")
plt.show()
```

```python
A = [[np.random.randint(1, 5) + np.random.random(1)[0], np.random.randint(3, 7) + np.random.random(1)[0]] for x in range(20)]
B = [[np.random.randint(7, 11)+ np.random.random(1)[0], np.random.randint(1, 5) + np.random.random(1)[0]] for x in range(20)]
A.extend(B)
X1 = np.array(A)
y1 = [0 for x in range(20)]
y1.extend([1 for x in range(20)])
Y1 = ['#008fd5' if v==1 else 'red' for v in y1]
clf1 = svm.SVC(kernel='linear', C=1000, gamma='auto')
clf1.fit(X1, y1)
```

```python

plt.figure(figsize=(12, 6))
plt.scatter(X1[:, 0], X1[:, 1], c=Y1, s=100)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf1.decision_function(xy).reshape(XX.shape)

ax.arrow(1, 1.1, 10, 0, length_includes_head = True, head_width = 0.2, head_length=.1, color='black' )
ax.arrow( 1, 1.1, 0, 6, length_includes_head = True, head_width = 0.15, head_length=.18, color='black' )


ax.plot(3.5, 1.7, 'o', markersize=10)
ax.plot(np.linspace(2, 10, 100), .7*np.linspace(2, 10, 100)+.1, linewidth=1.5, color='m')


# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support vectors
#ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
#           linewidth=1, facecolors='none', edgecolors='k')

plt.savefig("LargeC1.eps")
```
