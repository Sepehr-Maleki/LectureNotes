```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)
%matplotlib inline
```

```python
class gridworld(object):
    def __init__(self):
        self.gridSize = 4
        self.terminal_states = [[0, 0], [self.gridSize-1, self.gridSize-1]]
        self.action_space = [[-1, 0], [1, 0], [0, 1], [0, -1]]
        self.gamma = 1 #Undiscounted
        self.actionProb = 0.25
        self.r = -1
        self.v = np.zeros((self.gridSize, self.gridSize))
        self.q = np.zeros((self.gridSize, self.gridSize))
        self.states = [[i, j] for i in range(self.gridSize) for j in range(self.gridSize)]
        
    def Reward(self,s, a):
        """
        s: initial position
        a: action
        """
        if s in self.terminal_states:
            s_prime = s
            r = 0
        else:
            r = self.r
            s_prime = np.array(s) + np.array(a)
            if -1 in s_prime or 4 in s_prime:
                s_prime = s
        return s_prime, r
    
    def evaluate(self, iterations=200):
        Delta = []
        for k in range(iterations):
            if k==0:
                table = plt.table(cellText=np.round(self.v, 1), colWidths=[.05] * 4, loc='center')
                table.auto_set_font_size(False)
                table.set_fontsize(20)
                table.scale(4, 4)
                
                plt.title("$v_{0}$".format(k))
                plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
                plt.tick_params(axis='y', which='both', right=False, left=False, labelleft=False)
                for pos in ['right','top','bottom','left']:
                    plt.gca().spines[pos].set_visible(False)
                plt.savefig('values_{0}.eps'.format(k), bbox_inches='tight', pad_inches=0.05)
                
            vNew = np.copy(self.v)
            for s in self.states:
                weightedRewards=0
                for action in self.action_space:
                    s_prime , r = self.Reward(s, action)
                    weightedRewards += self.actionProb * 1 * (r + (self.gamma * self.v[s_prime[0], s_prime[1]]))
                vNew[s[0], s[1]] = weightedRewards
                Delta.append(np.abs(self.v[s[0], s[1]] - weightedRewards))
                
            self.v = vNew
            
            if k in [0, 1, 2, 3, 9, iterations-1]:
                table = plt.table(cellText=np.round(self.v, 1), colWidths=[.05] * 4, loc='center')
                table.auto_set_font_size(False)
                table.set_fontsize(20)
                table.scale(self.gridSize, self.gridSize)
                if k == iterations - 1:
                    plt.title("$v_\pi$")
                elif k==9:
                    plt.title("$v_{10}$")
                else:
                    plt.title("$v_{0}$".format(k+1))
                    
                plt.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
                plt.tick_params(axis='y', which='both', right=False, left=False, labelleft=False)
                for pos in ['right','top','bottom','left']:
                    plt.gca().spines[pos].set_visible(False)
                #plt.savefig('values_{0}.eps'.format(k+1), bbox_inches='tight', pad_inches=0.05)

        plt.figure(figsize=(20, 10))
        plt.plot(Delta)        
        
    def policy(self, save=False):
        x = np.linspace(0, self.gridSize - 1, self.gridSize) + 0.5
        y = np.linspace(self.gridSize - 1, 0, self.gridSize) + 0.5
        X, Y = np.meshgrid(x, y)
        zeros = np.zeros((self.gridSize, self.gridSize))
        Up = np.zeros((self.gridSize, self.gridSize))
        Right = np.zeros((self.gridSize, self.gridSize))
        Down = np.zeros((self.gridSize, self.gridSize))
        Left = np.zeros((self.gridSize, self.gridSize))
        for state in self.states:
            if state not in self.terminal_states:
                neighbouring_states = [state + np.array(a) for a in self.action_space]
                neighbouring_states = [x for x in neighbouring_states if x[0]<4 and x[1]<4 and x[0] >=0 and x[1]>=0]
                neighbouring_values = [self.v[s[0], s[1]] for s in neighbouring_states]
                max_value = np.max(neighbouring_values)
                possible_ways = np.sum([x==max_value for x in neighbouring_values])
                max_value_ind = (np.where([np.round(x, 5)==np.round(max_value, 5) for x in neighbouring_values])[0]).tolist()
                max_value_neighbours = [neighbouring_states[y] for y in max_value_ind]
                
                # Find which actions takes the state to the neighbours with max values
                for n in max_value_neighbours:
                    action = n - state
                    if (action == [0, 1]).all():
                        Right[state[0], state[1]] = .4
                    if (action == [0, -1]).all():
                        Left[state[0], state[1]] = .4
                    if (action == [1, 0]).all():
                        Down[state[0], state[1]] = .4
                    if (action == [-1, 0]).all():
                        Up[state[0], state[1]] = .4

        # Plot the policy               
        fig = plt.figure(figsize=(12,8))
        ax = plt.axes()
        plt.quiver(X, Y,zeros,Up, scale=1, units='xy')
        plt.quiver(X, Y,zeros,-Down, scale=1, units='xy')
        plt.quiver(X, Y,Right,zeros, scale=1, units='xy')
        plt.quiver(X, Y,-Left,zeros, scale=1, units='xy')

        plt.xlim([0, 4])
        plt.ylim([0, 4])
        ax.xaxis.set_major_locator(MultipleLocator(1))
        ax.yaxis.set_major_locator(MultipleLocator(1))
        ax.xaxis.set_minor_locator(AutoMinorLocator(1))
        ax.yaxis.set_minor_locator(AutoMinorLocator(1))
        ax.set_yticklabels([])
        ax.set_xticklabels([])
        plt.grid(which='both', linestyle='--')
        if save:
            plt.savefig('policy.eps')
```

```python
G = gridworld()
```

```python
G.evaluate(iterations=500)
```

```python
G.policy(save=True)
```
